## 1. Pourquoi utiliser des fonctions ?

Une <b style="color:#e74c3c;">fonction</b>  est un ensemble d’ <b style="color:#e74c3c;">instructions</b>, dépendant généralement d’un ou de
plusieurs <b style="color:#e74c3c;">paramètres</b> , qui effectue une <b style="color:#e74c3c;">action</b>, et cette fonction, une fois définie,
peut être appelée depuis n’importe où dans le programme.


Utiliser des fonctions permet de :

- <b style="color:#e74c3c;">Eviter les répétitions</b> de code (principe DRY : *Don't Repeat Yourself*)
- <b style="color:#e74c3c;">Décomposer</b>un programme complexe en sous-programmes simples
- Mieux  <b style="color:#e74c3c;">Structurer</b>et rendre le code plus lisible
- <b style="color:#e74c3c;">Réutiliser</b> la même fonction dans différents programmes

### Exemple concret : sans vs avec fonction

```python
# --- SANS fonction : code répété 3 fois ---
print("Bonjour Alice !")
print("Bienvenue dans le cours.")
print("---")

print("Bonjour Bob !")
print("Bienvenue dans le cours.")
print("---")

print("Bonjour Charlie !")
print("Bienvenue dans le cours.")
print("---")
```

## Notion de renvoi (return)
### A- fonction cube 

```python
def cube(mon_nombre): 
    return mon_nombre**3
a = cube(2)
b = cube(5)
print(a+b)
```

La valeur renvoyée par la fonction  <b style="color:#e74c3c;">peut-être mémorisée</b>  par exemple:  **a = cube(2)**.  
La fonction peut <b style="color:#e74c3c;">être appelée plusieurs fois.</b> 

### B- fonction nombre premier

```python
def premier(nb):
    for d in range(2,nb):
        if nb % d == 0:
            return False
    return True

a = premier(7)
print(a)
```

```python
b = premier(4)
print(b)
```

Il peut y avoir **plusieurs** instructions **return** mais chacune interrompt l’exécution de la fonction !

### B- Avec renvoi (return)

```python
# --- AVEC fonction : code écrit une seule fois, appelé 3 fois ---
def accueil(prenom):
    # Retourne un message de bienvenue personnalisé
    return "Bonjour " + prenom 

print(accueil("Alice"))
print(accueil("Bob"))
print(accueil("Charlie"))
```

```python
a supprimer
# DEFINITION : Python met le code en mémoire, ne l'exécute pas encore
def compter_jusqu_a(nombre):
    for i in range(1, nombre + 1):
        print(i)

# APPEL : Python exécute maintenant le code de la fonction
compter_jusqu_a(5)
```

### 4.5 Le `return` interrompt la fonction

Dès qu'un `return` est atteint, la fonction s'arrête immédiatement.

```python
# Plusieurs return possibles — le premier atteint interrompt la fonction
def est_premier(nb):
    for d in range(2, nb):
        if nb % d == 0:
            return False   # divisible => pas premier, on sort immédiatement
    return True            # aucun diviseur trouvé => premier

print(est_premier(7))   # True
print(est_premier(4))   # False
print(est_premier(13))  # True
```

## 5. Espace et portée des variables (locale vs globale)

Une variable définie **dans** une fonction est **locale** : elle n'existe que pendant l'exécution de cette fonction.  
Une variable définie **hors** d'une fonction est **globale** : visible partout, mais une fonction ne peut pas la modifier.

<div style="text-align:left;">
  <img src="https://raw.githubusercontent.com/HjrManel/Cours_python/main/espace_et_portee_des_variables.png" width="600">
</div>

<div style="background-color:#f8d7da; border-left:5px solid #dc3545; padding:12px; border-radius:4px;">
<b>Piège classique</b> — Le <code>b</code> dans la fonction est une copie locale. 
Modifier le <code>b</code> local ne change pas le <code>b</code> global !
</div>

```python
def test():
    b = 5          # variable locale à la fonction
    print(a, b)    # voit la variable globale a

a = 2  # variable globale
b = 7  # variable globale (différente du b local dans test)

test()         # affiche : 2 5
print(a, b)    # affiche : 2 7  (le b global n'a pas changé)
```
