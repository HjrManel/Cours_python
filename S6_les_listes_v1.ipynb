## Rappel &mdash; Qu'est-ce qu'une liste ?

Une <b style="color:#e74c3c;">liste</b> est une <b style="color:#e74c3c;">structure de données</b> : une manière d'organiser une collection de données.<br>
En Python, la liste est une structure <b style="color:#e74c3c;">ordonnée</b>, dont chaque élément est accèssible par son <b style="color:#e74c3c;">indice</b>.<br><br>
Les éléments peuvent être de types différents. L'indice du premier élément est toujours <code>0</code>.

```python
# Creation et affichage d'une liste
villes = ["Geneve", "Nyon", "Morges", "Lausanne"]
print(villes)
print(type(villes))
```

## 1. Accès à un élément

### Accès par indice positif ou négatif

<div style="background-color:#e8f4f8; border-left:5px solid #306998; padding:15px; border-radius:6px; margin:10px 0;">
  <b style="color:#306998;">Principe :</b>
  <ul>
    <li>Indice positif : part du début &mdash; <code>liste[0]</code> est le premier élément</li>
    <li>Indice négatif : part de la fin &mdash; <code>liste[-1]</code> est le dernier élément</li>
  </ul>
</div>

```python
# Exemple d'acces par indice
villes = ["Geneve", "Nyon", "Morges", "Lausanne"]

premiere = villes[0]     # premier element
quatrieme = villes[3]    # quatrieme element
derniere = villes[-1]    # dernier element (indice negatif)

print("Premier :", premiere)
print("Quatrieme :", quatrieme)
print("Dernier (indice -1) :", derniere)
```

```python
# Modification d'un element a un indice donne
villes[1] = "Rolle"    # remplace "Nyon" par "Rolle"
print("Liste mise a jour :", villes)
```

## 2. Tranches (slices)

Une tranche copie une partie de la liste.<br>
Le début est <b style="color:#e74c3c;">inclus</b>, la fin est <b style="color:#e74c3c;">exclue</b>.

| Syntaxe    | Signification                                |
|------------|----------------------------------------------|
| `lst[:]`   | Copie complète de la liste                   |
| `lst[m:]`  | Du m-ième élément jusqu'à la fin             |
| `lst[:n]`  | Du début jusqu'àu n-ième élément exclu       |
| `lst[m:n]` | Du m-ième au n-ième élément (n exclu)        |

```python
# Exemple complet de tranches
jours = ["lu", "ma", "me", "je", "ve", "sa", "di"]

print(jours)          # liste complete
print(jours[3])       # element a l'indice 3
print(type(jours))    # <class 'list'>
print(type(jours[3])) # <class 'str'>

print(jours[2:5])     # indices 2, 3, 4 => ["me", "je", "ve"]
print(jours[3:])      # du jeudi jusqu'a la fin
print(jours[:4])      # du debut jusqu'au 4eme exclu
```

## 3. Ajouter un élément

- <b style="color:#e74c3c;">append(el)</b> : ajoute à la fin
- <b style="color:#e74c3c;">insert(pos, el)</b> : insère à une position donnée (les suivants sont décalés)

```python
# Exemple : ajouter des elements
liste = [2, 4, 6, 8, 10, 12]

liste.append(14)      # ajout en fin de liste
print("Apres append :", liste)

liste.insert(2, 999)  # insertion a l'indice 2
print("Apres insert :", liste)
```

```python
# Inserer entre deux elements existants
villes = ["Geneve", "Morges"]
villes.insert(1, "Nyon")   # insere "Nyon" entre "Geneve" et "Morges"
print(villes)
```

## 4. Supprimer un élément

- <b style="color:#e74c3c;">remove(el)</b> : supprime la première occurrence de la valeur
- <b style="color:#e74c3c;">pop(pos)</b> : supprime l'élément a la position donnée et le retourne

```python
# Exemple : supprimer des elements
jours = ["lu", "ma", "me", "je", "ve", "sa", "di"]

jours.remove("ve")       # supprime la valeur "ve"
print("Apres remove :", jours)

retire = jours.pop(1)    # supprime l'element a l'indice 1 et le memorise
print("Element retire :", retire)
print("Liste finale :", jours)
```

## 5. Trier une liste

La méthode <b style="color:#e74c3c;">sort()</b> trie la liste en place (la liste originale est modifiée).<br>
Le tri suit l'ordre <b style="color:#e74c3c;">ASCII</b> : majuscules avant minuscules, puis ordre alphabétique.<br>

<div style="background-color:#f8d7da; border-left:5px solid #dc3545; padding:12px; border-radius:4px;">
<b>Attention</b> — Tous les éléments doivent être du <b>même type</b> pour trier une liste.
</div>

```python
# Tri ascendant et descendant
fruits = ["poire", "banane", "abricot", "cerise"]

fruits.sort()             # tri alphabetique croissant
print("Croissant :", fruits)

fruits.sort(reverse=True) # tri alphabetique decroissant
print("Decroissant :", fruits)
```

## 6. Choisir un élément aléatoire : random.choice()

La fonction <b style="color:#e74c3c;">choice()</b> du module <code>random</code> retourne un seul élément choisi au hasard dans une liste.

```python
from random import choice

couleurs = ["rouge", "vert", "bleu"]
tirage = choice(couleurs)   # element aleatoire
print("Couleur tiree :", tirage)
```

```python
from random import choice

# Tirer 20 fruits et compter les occurrences
fruits = ["Banane", "Abricot", "Cerise", "Poire"]
compteur = {}

for _ in range(20):
    fruit = choice(fruits)
    compteur[fruit] = compteur.get(fruit, 0) + 1

print("Resultats apres 20 tirages :", compteur)
```

## 7. Liste de listes

Une liste peut contenir d'autres listes comme éléments.<br>
Pour accéder à un élément imbriqué, on chaîne les indices : <b style="color:#e74c3c;">liste[i][j]</b>

```python
# Exemple : liste de listes imbriquees
etats = [["chaud", "tiede", ["froid", "glace"]], ["lumineux", "eclaire", "sombre"]]

# "lumineux" : 2e tableau (ind 1), 1er element (ind 0)
e1 = etats[1][0]
print("lumineux :", e1)

# "glace" : 1er tableau (ind 0), 3e element (ind 2), 2e choix (ind 1)
e2 = etats[0][2][1]
print("glace :", e2)

# "froid" : 1er tableau (ind 0), 3e element (ind 2), 1er choix (ind 0)
e3 = etats[0][2][0]
print("froid :", e3)

# Modifier "sombre" par "fonce"
etats[1][2] = "fonce"
print("Liste modifiee :", etats)
```

## 8. Taille d'une liste : len()

<b style="color:#e74c3c;">len(liste)</b> retourne le nombre d'éléments de <b>premier niveau</b> de la liste.

```python
# Exemples de len()
jours = ["lu", "ma", "me", "je", "ve", "sa", "di"]
print("Taille jours :", len(jours))    # 7

vide = []
print("Taille vide :", len(vide))      # 0

# Seuls les elements de premier niveau sont comptes
multi = [["x","y","z"], [[1,2],[3,4]], [1,2,3], "a", 10]
print("Taille multi :", len(multi))    # 5
```

```python
# Verification sur des listes variees
fruits  = ["banane", "abricot", "cerise", "poire"]
nombres = [1, 2, 3, 4, 5]
etrange = [[], [[], []], [], [], [[], []]]

print("fruits :", len(fruits))    # 4
print("nombres :", len(nombres))  # 5
print("etrange :", len(etrange))  # 5
```

## 9. Opérateurs + et * sur les listes

<div style="background-color:#e8f4f8; border-left:5px solid #306998; padding:15px; border-radius:6px; margin:10px 0;">
  <b style="color:#306998;">À retenir :</b>
  <ul>
    <li><code>lst1 + lst2</code> : concatène (assemble) deux listes</li>
    <li><code>n * lst</code> : répète la liste n fois</li>
    <li><code>lst1 * lst2</code> : impossible &mdash; Python lève une erreur</li>
  </ul>
</div>

```python
# Operateurs + et * sur les listes
lst1 = [1, 2]
lst2 = [3, 4]

print("lst1 :", lst1)
print("lst2 :", lst2)

lst3 = lst1 + lst2       # concatenation
print("lst1 + lst2 :", lst3)

lst4 = 2 * lst1          # repetition
print("2 * lst1 :", lst4)

lst5 = 10 * []           # liste vide repetee => liste vide
print("10 * [] :", lst5)

lst6 = 10 * [0]          # dix zeros
print("10 * [0] :", lst6)

# lst7 = lst1 * lst2    # ERREUR : TypeError — decommentez pour verifier
# print(lst7)
```

### Exercice A &mdash; Parcourir une liste avec son indice

Afficher chaque élément d'une liste sur une ligne séparée, avec son indice.

```python
# Parcourir une liste avec son indice
langages = ["Python", "Java", "C", "JavaScript"]

for i in range(len(langages)):
    print(i, "->", langages[i])
```

### Exercice B &mdash; Test d'appartenance

Tester la présence ou l'absence d'un élément avec <code>in</code> / <code>not in</code>.

```python
# Test d'appartenance avec in / not in
courses = ["pain", "lait", "fromage", "beurre"]

if "lait" in courses:
    print("Le lait est dans la liste")

if "jus" not in courses:
    print("Le jus n'est pas dans la liste")
```

### Exercice C &mdash; Construire une liste par accumulation

Demander des notes à l'utilisateur, les stocker dans une liste, puis calculer la moyenne.

```python
# Construction d'une liste par accumulation
notes = []
n = int(input("Combien de notes ? "))

for i in range(n):
    note = float(input("Note " + str(i+1) + " : "))
    notes.append(note)

# sum() calcule la somme des elements
moyenne = sum(notes) / len(notes)
print("Notes :", notes)
print("Moyenne :", moyenne)
```

### Exercice D &mdash; Nettoyer une liste

Construire une nouvelle liste en retirant tous les zéros d'une liste d'origine.

```python
# Retirer tous les zeros d'une liste
donnees = [3, 0, 7, 0, 1, 0, 9, 4]
propres = []

for val in donnees:
    if val != 0:
        propres.append(val)

print("Original :", donnees)
print("Nettoyee :", propres)
```
