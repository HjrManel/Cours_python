**Comment organiser un tel dessin ?**

Vous pourriez écrire un énorme bloc de code qui dessine tout d'un coup. Mais ce serait difficile à lire, à modifier et à réutiliser. Une bien meilleure idée : **découper le problème en petits morceaux**.

- Une fonction `porte()` qui dessine seulement la porte
- Une fonction `fenetre()` qui dessine seulement la fenêtre
- Une fonction `maison()` qui **appelle** `porte()` et `fenetre()` au bon endroit

C'est exactement comme en architecture : l'architecte conçoit la maison dans son ensemble, mais utilise les compétences d'un menuisier pour la porte et d'un vitrier pour la fenêtre. La fonction `maison()` est comme l'architecte, `porte()` et `fenetre()` sont ses collaborateurs.

## Une fonction peut en appeler une autre

Commençons simplement. Observez attentivement la cellule de code ci-dessous.

Elle contient **deux définitions de fonctions** :
- `porte()` qui dessine un petit rectangle (la porte)
- `maison()` qui dessine les murs ET qui **appelle `porte()` à l'intérieur** de sa propre définition

### Deux règles importantes à comprendre avant d'exécuter

1. **L'ordre compte** : la fonction `porte()` doit être définie **AVANT** `maison()`. Sinon, quand Python lit la définition de `maison()` et rencontre l'appel à `porte()`, il ne sait pas encore ce qu'est `porte()`.

2. **L'indentation indique l'intérieur** : l'appel à `porte()` est *indenté* (décalé vers la droite) car il fait partie du bloc de `maison()`. C'est cette indentation qui fait que `porte()` est *à l'intérieur* de `maison()`.


```python
# Exemple commente : porte() est appelee a l'interieur de maison()
from turtle import *

# Decalage pour centrer la maison dans la fenetre
penup()
goto(-100, -50)
setheading(0)
pendown()

# ETAPE 1 : on definit d'abord la fonction porte()
def porte():
    # Un petit rectangle de 20 de large par 40 de haut
    for _ in range(2):
        forward(20)
        left(90)
        forward(40)
        left(90)

# ETAPE 2 : on definit ENSUITE la fonction maison()
# qui utilise porte() a l'interieur
def maison():
    # D'abord, on dessine les 4 murs (rectangle 200 x 100)
    for _ in range(2):
        forward(200)
        left(90)
        forward(100)
        left(90)
    # Puis on va a l'emplacement precis de la porte (en bas, centree)
    penup()
    goto(-10, -50)            # coin bas gauche de la porte
    setheading(0)
    pendown()
    # Et on appelle porte() : cela dessine la porte a cet endroit
    porte()

# ETAPE 3 : on appelle maison() UNE SEULE fois
# Cet unique appel va declencher : les 4 murs PUIS la porte
maison()

done()
```

<div style="padding:18px; background-color:#e8f5e9; border:1px solid #66bb6a; border-left:6px solid #2e7d32; border-radius:4px;">
<h3 style="margin-top:0; color:#1b5e20;">Ce qu'il faut retenir</h3>

Quand vous écrivez `porte()` **à l'intérieur** de la définition de `maison()`, cela signifie :

> *« Quand on appelle `maison()`, à ce moment précis, fais aussi ce que fait `porte()`. »*

Avantages de cette organisation :

- **Lisibilité** : le code de `maison()` est court et parle de ce qu'il fait ("mur puis porte"), pas des détails techniques.
- **Réutilisation** : si on veut une autre maison ailleurs, on rappelle `maison()` une seule fois, et tout vient avec.
- **Modification facile** : si on change la forme de la porte, on modifie `porte()` une seule fois, et **toutes les maisons** qui utilisent `porte()` ont automatiquement la nouvelle porte.

### Défi bonus (pour les plus rapides)

Ajoutez un **toit triangulaire** à votre maison avec une fonction `toit()` qui dessine un triangle.

*Astuce : la tortue doit se positionner en haut à gauche du rectangle des murs avant de dessiner le toit.*


```python
# Exercice - solution
from turtle import *

# Decalage pour centrer la maison
penup()
goto(-100, -50)
setheading(0)                     # on remet la tortue a l'horizontale
pendown()

# Fonction porte : un rectangle 20 x 40
def porte():
    for _ in range(2):
        forward(20)
        left(90)
        forward(40)
        left(90)

# Fonction fenetre : un carre de cote 30
def fenetre():
    for _ in range(4):
        forward(30)
        left(90)

# Fonction maison qui appelle porte() ET fenetre() a des positions absolues
def maison():
    # Les 4 murs : rectangle de (-100, -50) a (100, 50)
    for _ in range(2):
        forward(200)
        left(90)
        forward(100)
        left(90)

    # Porte : positionnee en bas au milieu, sur le mur du bas
    # La porte fait 20 de large et 40 de haut, on la place entre x=-10 et x=10
    penup()
    goto(-10, -50)                # coin bas gauche de la porte
    setheading(0)                 # horizontale
    pendown()
    porte()

    # Fenetre : en haut a droite, un carre de 30
    # On la place entre x=30 et x=60, y=0 a y=30
    penup()
    goto(30, 0)                   # coin bas gauche de la fenetre
    setheading(0)                 # horizontale
    pendown()
    fenetre()

# Un seul appel a maison() dessine tout : murs + porte + fenetre
maison()

done()
```
