## 1. Pourquoi utiliser des fonctions ?

Une **fonction** est un ensemble d'**instructions**, dépendant généralement d'un ou de
plusieurs **paramètres**, qui effectue une **action**. Une fois définie,
elle peut être appelée depuis n'importe où dans le programme.

Utiliser des fonctions permet de :

- **Éviter les répétitions** de code (principe DRY : *Don't Repeat Yourself*)
- **Décomposer** un programme complexe en sous-programmes simples
- Mieux **structurer** et rendre le code plus lisible
- **Réutiliser** la même fonction dans différents programmes

### Exemple : sans vs avec fonction

````python
# --- SANS fonction : le message de bienvenue est répété 3 fois ---
print("Bonjour Alice !")
print("Bienvenue dans le cours.")
print("---")

print("Bonjour Bob !")
print("Bienvenue dans le cours.")
print("---")

print("Bonjour Charlie !")
print("Bienvenue dans le cours.")
print("---")
````

````python
# --- AVEC fonction : le code est écrit une seule fois, appelé 3 fois ---
def accueil(prenom):
    # Affiche un message de bienvenue personnalisé
    print("Bonjour", prenom)
    print("Bienvenue dans le cours.")
    print("---")

accueil("Alice")
accueil("Bob")
accueil("Charlie")
````

| Fonction | Retourne quelque chose ? | Exemple |
|----------|--------------------------|--------|
| `print()` | Non | `print("Bonjour")` |

### 4.5 Le `return` interrompt la fonction

Dès qu'un `return` est atteint, la fonction s'arrête **immédiatement**.
Il peut y avoir plusieurs `return`, mais un seul sera exécuté.

````python
# Plusieurs return possibles — le premier atteint interrompt la fonction
def est_premier(nb):
    for d in range(2, nb):
        if nb % d == 0:
            return False   # divisible => pas premier, on sort immédiatement
    return True            # aucun diviseur trouvé => premier

print(est_premier(7))    # True
print(est_premier(4))    # False
print(est_premier(13))   # True
````

## 5. Portée des variables : locale vs globale

Une variable définie **dans** une fonction est **locale** : elle n'existe que pendant
l'exécution de cette fonction.  
Une variable définie **hors** d'une fonction est **globale** : visible à l'intérieur,
mais la fonction **ne peut pas la modifier**.

<div style="text-align:left;">
  <img src="https://raw.githubusercontent.com/HjrManel/Cours_python/main/espace_et_portee_des_variables.png" width="600">
</div>

````python
def test():
    b = 5          # variable locale à la fonction
    print(a, b)    # voit la variable globale a

a = 2  # variable globale
b = 7  # variable globale (distincte du b local dans test)

test()         # affiche : 2 5
print(a, b)    # affiche : 2 7  — le b global n'a pas changé
````

<div style="background-color:#f8d7da; border-left:5px solid #dc3545; padding:12px 16px; border-radius:4px; margin:12px 0;">
<b>Attention :</b> Le <code>b</code> dans la fonction est une copie <strong>locale</strong>. Modifier <code>b</code> à l'intérieur de <code>test()</code> ne change pas le <code>b</code> global !
</div>

````python
# Vérification : la variable locale disparaît après la fonction
def ma_fonction():
    x_local = 99
    print("Dans la fonction, x_local =", x_local)

ma_fonction()

# Décommenter la ligne suivante pour voir l'erreur :
# print(x_local)  # NameError : x_local n'existe pas ici
````

<div style="background-color:#f0eafa; border-left:5px solid #8e44ad; padding:14px 18px; border-radius:4px; margin:14px 0;">
<b>Application — Exemple complet a imiter — Parcourir une chaine</b><br>
Cette fonction compte les occurrences de la lettre <code>'a'</code> dans un mot.<br>
Repérez : l'accumulateur, la boucle sur la chaine, la condition, le return.
</div>

````python
# Exemple complet : compter les "a" dans un mot
def compter_a(mot):
    compteur = 0              # accumulateur initialise a 0
    for lettre in mot:        # on parcourt chaque lettre
        if lettre == "a":     # condition
            compteur = compteur + 1
    return compteur

print(compter_a("ananas"))       # affiche : 3
print(compter_a("informatique")) # affiche : 1
print(compter_a("python"))       # affiche : 0
````
