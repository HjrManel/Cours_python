**Géométrie d'un bungalow :** un carré de côté 80 surmonté d'un toit en triangle équilatéral de côté 80. Le tracé se fait **d'un seul trait continu**, en faisant le tour du contour complet (bas, côté droit, toit droit, toit gauche, côté gauche).


```python
# Exercice 3 - Partie 1 : FACTORISER ce code en utilisant une fonction bungalow()
from turtle import *

# Decalage pour commencer a gauche de la fenetre
up()
goto(-100, -60)
down()

# --- Premier bungalow (dessine "a la main") ---
forward(80)           # cote bas
left(90)
forward(80)           # cote droit
left(30)
forward(80)           # toit, cote droit
left(120)
forward(80)           # toit, cote gauche
left(30)
forward(80)           # cote gauche (redescente)
left(90)

# Deplacement vers le deuxieme bungalow
up()
forward(100)
down()

# --- Deuxieme bungalow (meme code, repete a l'identique !) ---
forward(80)
left(90)
forward(80)
left(30)
forward(80)
left(120)
forward(80)
left(30)
forward(80)
left(90)

done()
```
