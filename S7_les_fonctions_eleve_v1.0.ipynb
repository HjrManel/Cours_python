## 2. Nouvelle structure :

Serait-il possible de **donner un nom** aux 8 lignes du rectangle, pour ne les écrire qu'une seule fois et les réutiliser autant que l'on veut ?

**--> Oui**, grâce au mot-clé `def`. Vous venez de débloquer un nouveau **super-pouvoir de développeur·euse** : les **fonctions**.


````python
# Avec fonction : le code du rectangle est ecrit UNE SEULE fois
from turtle import *

# Decalage du point de depart vers la gauche
up()
goto(-250, 0)
down()

# On regroupe les 8 lignes du rectangle sous un nom
def rectangle():
    forward(150)
    left(90)
    forward(90)
    left(90)
    forward(150)
    left(90)
    forward(90)
    left(90)

rectangle() # appel de la fonction 

# Deplacement entre les deux rectangles
up()
forward(200)
down()

rectangle() # appel de la fonction pour dessiner le 2ème rectangle

done()
````

<div style="padding:18px; background-color:papayawhip; border:1px solid #f0b37e; border-left:6px solid #e07a20; border-radius:4px;">
<h2 style="margin-top:0">Notions importantes à retenir</h2>


Avec le mot-clé `def`, on peut **définir** une nouvelle commande (ici `rectangle()`). On dit qu'on définit une **fonction**.

Le code qui suit `def rectangle():` doit être en **indentation** (décalé de 4 espaces généralement ou une tabulation vers la droite).

Ensuite, il suffit d'écrire `rectangle()` pour exécuter le code du rectangle. On dit qu'on **appelle** la fonction.

**Retenez bien :**
- On **définit** une fonction **une seule fois**,
- On **appelle** une fonction **autant de fois que l'on veut**,
- Si on ne l'appelle pas, elle n'est **pas exécutée** et rien ne se passe à l'écran.
</div>

## 3. Définition et anatomie d'une fonction

<div style="border: 1px solid #d1d5db; border-radius: 8px; padding: 20px 24px; background-color: #ffffff; font-family: Arial, sans-serif; margin: 20px 0;">

<h3 style="font-size: 18px; font-weight: 500; margin: 0 0 12px 0; border-left: 4px solid #2c3e50; padding-left: 10px; color: #2c3e50;">Définition</h3>

<p style="font-size: 15px; line-height: 1.6; margin: 0 0 10px 0; color: #2c3e50;">Une fonction est un <span style="color:#c0392b; font-weight:500;">ensemble d'instructions</span> qui effectue une <span style="color:#c0392b; font-weight:500;">action</span>. Une fois définie, elle peut être appelée depuis n'importe où dans le programme.</p>

<p style="font-size: 15px; line-height: 1.6; margin: 14px 0 8px 0; color: #2c3e50;">Écrire une fonction permet :</p>

<ul style="margin: 0; padding-left: 8px; list-style: none;">
  <li style="font-size: 15px; line-height: 1.8; padding-left: 22px; position: relative; color: #2c3e50;"><span style="position: absolute; left: 0; color: #7f8c8d;">&rarr;</span> <span style="color:#c0392b; font-weight:500;">d'éviter</span> les répétitions de code</li>
  <li style="font-size: 15px; line-height: 1.8; padding-left: 22px; position: relative; color: #2c3e50;"><span style="position: absolute; left: 0; color: #7f8c8d;">&rarr;</span> de <span style="color:#c0392b; font-weight:500;">décomposer</span> un programme complexe en sous-programmes plus simples</li>
  <li style="font-size: 15px; line-height: 1.8; padding-left: 22px; position: relative; color: #2c3e50;"><span style="position: absolute; left: 0; color: #7f8c8d;">&rarr;</span> de mieux <span style="color:#c0392b; font-weight:500;">structurer</span> un programme</li>
  <li style="font-size: 15px; line-height: 1.8; padding-left: 22px; position: relative; color: #2c3e50;"><span style="position: absolute; left: 0; color: #7f8c8d;">&rarr;</span> la <span style="color:#c0392b; font-weight:500;">réutilisation</span> : une même fonction peut être utilisée dans différents programmes</li>
</ul>

<strong>Erreur fréquente n°1 :</strong> croire qu'écrire seulement <code>def rectangle():</code> dessine un rectangle.<br>
Tant qu'on n'a pas <strong>appelé</strong> la fonction (en écrivant <code>rectangle()</code>), rien ne se passe !

<div style="background-color:#fbd9dc; border-left:5px solid #e74c3c; padding:12px 16px; border-radius:6px; margin:16px 0;">
<strong>Erreur fréquente n°2 :</strong> oublier les parenthèses lors de l'appel.<br>
Écrire <code>rectangle</code> (sans les parenthèses) n'exécute pas la fonction ! Il faut bien écrire <code>rectangle()</code>.
</div>


<div style="background-color:#fbd9dc; border-left:5px solid #e74c3c; padding:12px 16px; border-radius:6px; margin:16px 0;">
<strong>Erreur fréquente n°3 :</strong> définir la fonction avec un nom et l'appeler avec un autre.<br>
Si on écrit <code>def rectangle():</code> puis <code>rectang()</code> (sans le <code>le</code> final), Python ne reconnaît pas <code>rectang</code> : il va afficher une erreur <code>NameError</code>. Le nom à la définition et à l'appel doit être <strong>strictement identique</strong>.
</div>
