## 1. Reprenons votre fonction `triangle()`

Voici la fonction que vous aviez écrite en partie 1. Elle dessine **un triangle équilatéral de côté 100**, toujours le même.

````python
from turtle import *

# La fonction triangle() de la partie 1 : cote fixe = 100
def triangle():
    for _ in range(3):
        forward(100)
        left(120)

# On l'appelle 3 fois pour voir
up()
goto(-250, 0)
down()

triangle()
up()
forward(150)
down()
triangle()
up()
forward(150)
down()
triangle()

done()
````

**Que constatez-vous ?** Les trois triangles ont **la même taille**. Mais le responsable en veut de tailles différentes.

### Première idée : créer trois fonctions

Si on a besoin de trois tailles, écrivons trois fonctions, une pour chaque taille :

````python
from turtle import *

# Tentative naive : une fonction par taille
def triangle_petit():
    for _ in range(3):
        forward(50)
        left(120)

def triangle_moyen():
    for _ in range(3):
        forward(100)
        left(120)

def triangle_grand():
    for _ in range(3):
        forward(150)
        left(120)

# On les appelle alignes
up()
goto(-250, 0)
down()

triangle_petit()
up()
forward(100)
down()
triangle_moyen()
up()
forward(150)
down()
triangle_grand()

done()
````

## 2. La solution : donner un **paramètre** à la fonction

Au lieu d'écrire la taille **en dur** dans le code (50, 100, 150...), on écrit un **paramètre** entre les parenthèses de `def`. Sa valeur sera donnée au moment de l'appel

````python
from turtle import *

# Une seule definition. Le parametre 'cote' represente la taille.
# Sa valeur sera donnee au moment de l'appel.
def triangle(cote):
    for _ in range(3):
        forward(cote)
        left(120)

# Trois appels, trois tailles differentes
up()
goto(-250, 0)
down()

triangle(50)
up()
forward(100)
down()

triangle(100)
up()
forward(140)
down()

triangle(150)

done()
````

<h2 style="margin-top:0">Notion importante : paramètre et argument</h2>

<h3 style="margin-top:0; color:#854F0B;">Comparons les deux écritures</h3>

```
def triangle(cote):       <- définition
    forward(cote)
    left(120)

triangle(100)             <- appel
```

1. Le mot `cote` apparaît dans la **définition** de la fonction, entre les parenthèses. On l'appelle un **paramètre**. C'est un nom que **vous choisissez** : il pourrait s'appeler `taille`, `c`, `longueur`... ce qui compte, c'est qu'il soit lisible et qu'il ait un sens. Même règle d'appellation que les variables.

2. Le nombre `100` apparaît dans **l'appel** de la fonction. On l'appelle un **argument**. C'est **la valeur** que la fonction va utiliser pour faire son travail.

3. La grande astuce du paramètre : **à chaque appel**, sa valeur peut être différente.
   - Avec `triangle(50)`, le paramètre `cote` vaut 50.
   - Avec `triangle(100)`, il vaut 100.
   - Avec `triangle(200)`, il vaut 200.

4. La définition est écrite **une seule fois**. Pourtant, elle peut produire **autant de dessins différents** qu'il y a d'appels avec des valeurs différentes. C'est exactement le contraire de ce qu'on faisait juste avant avec `triangle_petit()`, `triangle_moyen()`, `triangle_grand()`.

5. Le paramètre n'existe **que dans la fonction**. Si vous tapez `print()` après la fonction, Python répond qu'il ne connaît pas `cote`. Le paramètre est interne, comme un mot de code qu'on partage entre la définition et chaque appel.

````python
# Exercice 1 - SOLUTION : 5 dalles carrees alignees, tailles croissantes
from turtle import *

# Position de depart : a gauche du jardin
up()
goto(-280, 0)
down()

# Etape 1 : definition de la fonction carre(cote)
# Le parametre 'cote' permettra de dessiner un carre de n'importe quelle taille
def carre(cote):
    for _ in range(4):
        forward(cote)
        left(90)

# Etape 2 : dessiner les 5 dalles avec une boucle for
# Chaque appel utilise une taille differente : le parametre s'adapte
for taille in [40, 60, 80, 100, 120]:
    carre(taille)
    up()
    forward(taille + 15)   # avancer de la largeur de la dalle + un petit espace
    down()

done()
````

## 4. Plusieurs paramètres dans une même fonction

Certaines formes ont besoin de **deux mesures** pour être complètement décrites. Un **rectangle**, par exemple : il a une largeur et une hauteur.

Python permet de mettre **autant de paramètres que nécessaire** dans la définition d'une fonction. Il suffit de les séparer par des **virgules**.

````python
from turtle import *

# Deux parametres : largeur et hauteur, separes par une virgule
def rectangle(largeur, hauteur):
    for _ in range(2):
        forward(largeur)
        left(90)
        forward(hauteur)
        left(90)

# Trois rectangles aux proportions differentes
# A chaque appel, l'ordre est respecte : largeur d'abord, hauteur ensuite
up()
goto(-260, -50)
down()

rectangle(100, 60)    # batiment large et bas
up()
forward(130)
down()

rectangle(60, 120)    # tour haute et etroite
up()
forward(90)
down()

rectangle(80, 80)     # batiment carre

done()
````

<strong>Erreur fréquente n°4 :</strong> Inverser l'ordre des arguments.<br>
Si vous voulez un rectangle de 200 de large et 50 de haut, vous devez écrire <code>rectangle(200, 50)</code> et non pas <code>rectangle(50, 200)</code>.
````python
# Exercice 2 - SOLUTION : 3 piscines bleues, dessin + aire
from turtle import *

# Position de depart : a gauche pour aligner les 3 piscines
up()
goto(-280, 0)
down()

# Definition de la fonction piscine(largeur, longueur)
# Elle dessine un rectangle bleu rempli ET affiche l'aire
def piscine(largeur, longueur):
    # Dessin du rectangle rempli
    fillcolor('lightblue')
    begin_fill()
    for _ in range(2):
        forward(largeur)
        left(90)
        forward(longueur)
        left(90)
    end_fill()
    
    # Calcul et affichage de l'aire (1 pixel = 1 m, donc en m2)
    aire = largeur * longueur
    print('Piscine', largeur, 'x', longueur, '-> aire :', aire, 'm2')

# Test avec les 3 dimensions demandees
piscine(80, 40)     # petite
up()
forward(100)
down()

piscine(120, 60)    # moyenne
up()
forward(140)
down()

piscine(180, 80)    # grande

done()
````
