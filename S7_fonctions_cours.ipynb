## 1. Pourquoi utiliser des fonctions ?

Une **fonction** est un bloc d'instructions réutilisable qui effectue une tâche précise.

Utiliser des fonctions permet de :

- **Eviter les répétitions** de code (principe DRY : *Don't Repeat Yourself*)
- **Décomposer** un programme complexe en sous-programmes simples
- **Structurer** et rendre le code plus lisible
- **Réutiliser** le même code dans différents programmes

### Exemple concret : sans vs avec fonction

````python
# --- SANS fonction : code répété 3 fois ---
print("Bonjour Alice !")
print("Bienvenue dans le cours.")
print("---")

print("Bonjour Bob !")
print("Bienvenue dans le cours.")
print("---")

print("Bonjour Charlie !")
print("Bienvenue dans le cours.")
print("---")
````

````python
# --- AVEC fonction : code écrit une seule fois, appelé 3 fois ---
def accueil(prenom):
    # Affiche un message de bienvenue personnalisé
    print("Bonjour", prenom, "!")
    print("Bienvenue dans le cours.")
    print("---")

accueil("Alice")
accueil("Bob")
accueil("Charlie")
````

## 3. Définir ses propres fonctions

### Syntaxe générale

```python
def nom_fonction(param1, param2):
    instruction...
    instruction...
    return valeur
```

Points importants :
- Le mot-clé `def` annonce la définition
- Le nom suit les mêmes règles que les variables (minuscules, pas d'espaces)
- Les `:` sont obligatoires après la parenthèse
- Le corps est **indenté** (4 espaces ou Tab)
- `return` est optionnel — une fonction sans `return` ne renvoie rien

### Différence : définition vs appel

````python
# DEFINITION : Python met le code en mémoire, ne l'exécute pas encore
def compter_jusqu_a(nombre):
    for i in range(1, nombre + 1):
        print(i)

# APPEL : Python exécute maintenant le code de la fonction
compter_jusqu_a(5)
````

<div style="background-color:#d1ecf1; border-left:5px solid #17a2b8; padding:12px; border-radius:4px;">
<b>A retenir :</b> La définition d'une fonction ne fait rien toute seule. 
Il faut <strong>appeler</strong> la fonction pour que le code s'exécute.

## 4. Les quatre types de fonctions

### 4.1 Sans paramètre, sans retour

````python
# Fonction sans paramètre, sans return
def compteur3():
    i = 0
    while i < 3:
        print(i)
        i = i + 1

compteur3()  # Appel — les parenthèses sont obligatoires !
````

### 4.2 Sans paramètre, avec retour

````python
# Fonction sans paramètre, avec return
def trois_au_cube():
    return 3 ** 3

resultat = trois_au_cube()
print(resultat)
````

### 4.3 Avec paramètre(s), avec retour

````python
# Fonction avec un paramètre
def cube(nombre):
    return nombre ** 3

a = cube(2)
b = cube(5)
print("2^3 =", a)
print("5^3 =", b)
print("Somme =", a + b)  # on peut utiliser directement le résultat
````

````python
# Fonction avec plusieurs paramètres
def puissance(nombre, exposant):
    return nombre ** exposant

print(puissance(2, 3))   # 8
print(puissance(10, 4))  # 10000
````

### 4.4 Le `return` interrompt la fonction

Dès qu'un `return` est atteint, la fonction s'arrête immédiatement.

````python
# Plusieurs return possibles — le premier atteint interrompt la fonction
def est_premier(nb):
    for d in range(2, nb):
        if nb % d == 0:
            return False   # divisible => pas premier, on sort immédiatement
    return True            # aucun diviseur trouvé => premier

print(est_premier(7))   # True
print(est_premier(4))   # False
print(est_premier(13))  # True
````

## 5. Portée des variables (locale vs globale)

Une variable définie **dans** une fonction est **locale** : elle n'existe que pendant l'exécution de cette fonction.  
Une variable définie **hors** d'une fonction est **globale** : visible partout, mais une fonction ne peut pas la modifier.

````python
def test():
    b = 5          # variable locale à la fonction
    print(a, b)    # voit la variable globale a

a = 2  # variable globale
b = 7  # variable globale (différente du b local dans test)

test()         # affiche : 2 5
print(a, b)    # affiche : 2 7  (le b global n'a pas changé)
````

<div style="background-color:#f8d7da; border-left:5px solid #dc3545; padding:12px; border-radius:4px;">
<b>Piège classique</b> — Le <code>b</code> dans la fonction est une copie locale. 
Modifier le <code>b</code> local ne change pas le <code>b</code> global !
</div>

````python
# Vérification : la variable locale disparait après la fonction
def ma_fonction():
    x_local = 99
    print("Dans la fonction, x_local =", x_local)

ma_fonction()

# Décommenter la ligne suivante pour voir l'erreur :
# print(x_local)  # NameError : x_local n'existe pas ici
````
