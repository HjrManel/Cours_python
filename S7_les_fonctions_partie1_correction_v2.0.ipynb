## 1. Pourquoi utiliser des fonctions ?

Votre responsable vous demande de dessiner **3 maisons identiques** pour commencer (en partant du plus simple : un rectangle par maison).

Voici le code qu'on pourrait écrire **sans fonction**. **Exécutez** la cellule ci-dessous pour voir le résultat.


````python
# Sans fonction : le code du rectangle est ecrit TROIS fois
from turtle import *

# Decalage du point de depart vers la gauche pour tout afficher
penup()
goto(-300, 0)
pendown()

# Rectangle 1 (longueur = 160, largeur = 100)
forward(160)
left(90)
forward(100)
left(90)
forward(160)
left(90)
forward(100)
left(90)

# Deplacement vers le 2e rectangle
penup()
forward(200)
pendown()

# Rectangle 2 (meme code, repete !)
forward(160)
left(90)
forward(100)
left(90)
forward(160)
left(90)
forward(100)
left(90)

# Deplacement vers le 3e rectangle
penup()
forward(200)
pendown()

# Rectangle 3 (encore une fois !)
forward(160)
left(90)
forward(100)
left(90)
forward(160)
left(90)
forward(100)
left(90)

done()
````

## 2. Nommer une séquence avec `def`

Serait-il possible de **donner un nom** aux 8 lignes du rectangle, pour ne les écrire qu'une seule fois et les réutiliser autant que l'on veut ?

**Oui**, grâce au mot-clé `def`.


````python
# Avec fonction : le code du rectangle est ecrit UNE SEULE fois
from turtle import *

# Decalage du point de depart vers la gauche
penup()
goto(-250, 0)
pendown()

# DEFINITION de la fonction (on lui donne un nom : 'rectangle')
def rectangle():
    forward(160)
    left(90)
    forward(100)
    left(90)
    forward(160)
    left(90)
    forward(100)
    left(90)

# APPEL de la fonction : on execute le code du rectangle
rectangle()           # premier appel -> dessine un rectangle

# Deplacement entre les deux rectangles
penup()
forward(200)
pendown()

rectangle()           # deuxieme appel -> dessine un second rectangle

done()
````

<div style="padding:18px; background-color:papayawhip; border:1px solid #f0b37e; border-left:6px solid #e07a20; border-radius:4px;">
<h2 style="margin-top:0">Notions importantes à retenir</h2>

Avec le mot-clé `def`, on peut **définir** une nouvelle commande (ici `rectangle()`). On dit qu'on définit une **fonction**.

Le code qui suit `def rectangle():` doit être en **indentation** (décalé de 4 espaces ou une tabulation vers la droite).

Ensuite, il suffit d'écrire `rectangle()` pour exécuter le code du rectangle. On dit qu'on **appelle** la fonction.

**Retenez bien :**
- On **définit** une fonction **une seule fois**,
- On **appelle** une fonction **autant de fois que l'on veut**,
- Si on ne l'appelle pas, elle n'est **pas exécutée** et rien ne se passe à l'écran.
</div>


<div style="background-color:#f8d7da; border-left:5px solid #dc3545; padding:12px 16px; border-radius:6px; margin:16px 0;">
<strong>Erreur fréquente n°1 :</strong> croire qu'écrire seulement <code>def rectangle():</code> suffit à dessiner un rectangle.<br>
Tant qu'on n'a pas <strong>appelé</strong> la fonction (en écrivant <code>rectangle()</code>), rien ne se passe !
</div>


<div style="background-color:#f8d7da; border-left:5px solid #dc3545; padding:12px 16px; border-radius:6px; margin:16px 0;">
<strong>Erreur fréquente n°2 :</strong> oublier les parenthèses lors de l'appel.<br>
Écrire <code>rectangle</code> (sans les parenthèses) n'exécute pas la fonction ! Il faut bien écrire <code>rectangle()</code>.
</div>

<h3 style="color:teal;background-color:azure; padding:8px 12px; border-radius:4px;"> &nbsp; Exercice 1 — Panneau de signalisation du chantier </h3>

*Votre responsable vous demande de dessiner un **panneau de signalisation triangulaire** pour baliser le chantier.*

1. Définissez une fonction `triangle()` qui dessine un **triangle équilatéral de côté 100**.
2. Appelez votre fonction **3 fois** avec un `left(120)` entre chaque appel pour former une rosace (un triptyque de panneaux).

**Indice :** un triangle équilatéral a 3 côtés égaux. La tortue tourne de **120°** entre chaque côté *(pourquoi 120° et pas 60° ? parce que la tortue tourne de son angle extérieur, pas intérieur)*.


````python
# Exercice 1 - solution
from turtle import *

# Definition : la fonction dessine UN triangle equilateral
def triangle():
    for _ in range(3):          # on repete 3 fois (3 cotes)
        forward(100)             # avance d'un cote
        left(120)                # tourne de 120 degres (angle exterieur)

# Appel : on dessine 6 triangles en tournant de 60 degres entre chaque
# -> cela forme une belle rosace (360 / 6 = 60)
triangle()
left(60)
triangle()
left(60)
triangle()
left(60)
triangle()
left(60)
triangle()
left(60)
triangle()

done()
````

````python
# Espace de test - experimentez !
# Modifiez les valeurs et observez le resultat :
# - Remplacez forward(100) par forward(50) ou forward(150) dans la fonction
# - Changez left(120) en right(120) -> que se passe-t-il ?
# - Appelez triangle() 6 fois au lieu de 3 (avec left(60) entre chaque)

from turtle import *

def triangle():
    for _ in range(3):
        forward(100)
        left(120)

triangle()

done()
````

## 3. Anatomie d'une fonction

Une **définition de fonction** comporte toujours les mêmes éléments. Regardez bien :

```python
def rectangle():
    forward(160)
    left(90)
    ...
```

| Élément | Rôle |
|:---|:---|
| 1. Le mot-clé **`def`** | Indique qu'on définit une fonction |
| 2. Le **nom** (`rectangle`) | Le nom qu'on choisit (en minuscules, sans espaces) |
| 3. Les **parenthèses** `()` | Obligatoires, même vides |
| 4. Les **deux-points** `:` | À la fin de la ligne `def` |
| 5. Le **bloc indenté** | Décalé de 4 espaces (le code de la fonction) |

**L'indentation est cruciale en Python** : c'est elle qui indique quelles lignes font partie de la fonction. Sans indentation, Python ne sait pas où s'arrête la fonction.

<h3 style="color:teal;background-color:azure; padding:8px 12px; border-radius:4px;"> &nbsp; Exercice 2 — Véhicule de chantier </h3>

*Pour votre plan du lotissement, vous devez également représenter un **véhicule de chantier** sur le site. Dessinons-le simplement : un **rectangle** (la carrosserie) et **deux cercles** (les roues).*

1. **Définissez** une fonction `carrosserie()` qui dessine un rectangle de 200 × 80.
2. **Définissez** une fonction `roue()` qui dessine un cercle de rayon 25.
3. **Appelez** vos deux fonctions pour dessiner le véhicule complet : une carrosserie, puis 2 roues placées sous la carrosserie.

**Indice pour placer les roues :** utilisez `penup()` et `pendown()` pour vous déplacer sans dessiner entre la carrosserie et la première roue, puis entre les deux roues.


````python
# Exercice 2 - solution
from turtle import *

# Decalage pour centrer la voiture dans la fenetre
penup()
goto(-100, 0)
pendown()

# Fonction 1 : la carrosserie = un rectangle 200 x 80
def carrosserie():
    for _ in range(2):           # 2 paires (longueur + largeur)
        forward(200)              # cote long
        left(90)
        forward(80)               # cote court
        left(90)

# Fonction 2 : une roue = un cercle de rayon 25
def roue():
    circle(25)

# On dessine la carrosserie : de (-100, 0) a (100, 80)
carrosserie()

# Roue 1 : position precise avec goto, sous la carrosserie a gauche
penup()
goto(-50, -25)                    # position absolue de la 1ere roue
setheading(0)                     # on remet la tortue a l'horizontale
pendown()
roue()

# Roue 2 : position precise avec goto, sous la carrosserie a droite
penup()
goto(50, -25)                     # position absolue de la 2eme roue
setheading(0)                     # on remet la tortue a l'horizontale
pendown()
roue()

done()
````
