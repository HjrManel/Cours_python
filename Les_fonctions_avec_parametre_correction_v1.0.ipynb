## 1. Reprenons votre fonction `triangle()`

Voici la fonction que vous aviez écrite en partie 1. Elle dessine **un triangle équilatéral de côté 100**, toujours le même.

````python
from turtle import *

# La fonction triangle() de la partie 1 : cote fixe = 100
def triangle():
    for _ in range(3):
        forward(100)
        left(120)

# On l'appelle 3 fois pour voir
up()
goto(-250, 0)
down()

triangle()
up()
forward(150)
down()
triangle()
up()
forward(150)
down()
triangle()

done()
````

**Que constatez-vous ?** Les trois triangles ont **la même taille**. Mais le responsable en veut de tailles différentes.

### Première idée : créer trois fonctions

Si on a besoin de trois tailles, écrivons trois fonctions, une pour chaque taille :

````python
from turtle import *

# Tentative naive : une fonction par taille
def triangle_petit():
    for _ in range(3):
        forward(50)
        left(120)

def triangle_moyen():
    for _ in range(3):
        forward(100)
        left(120)

def triangle_grand():
    for _ in range(3):
        forward(150)
        left(120)

# On les appelle alignes
up()
goto(-250, 0)
down()

triangle_petit()
up()
forward(100)
down()
triangle_moyen()
up()
forward(150)
down()
triangle_grand()

done()
````

## 2. La solution : donner un **paramètre** à la fonction

Au lieu d'écrire la taille **en dur** dans le code (50, 100, 150...), on écrit un **nom** entre les parenthèses de `def`. Ce nom est un **paramètre** : c'est une variable que la fonction recevra à chaque appel.

````python
from turtle import *

# Une seule definition. Le parametre 'cote' represente la taille.
# Sa valeur sera donnee au moment de l'appel.
def triangle(cote):
    for _ in range(3):
        forward(cote)
        left(120)

# Trois appels, trois tailles differentes
up()
goto(-250, 0)
down()

triangle(50)
up()
forward(100)
down()

triangle(100)
up()
forward(140)
down()

triangle(150)

done()
````

<h2 style="margin-top:0">Notion importante : paramètre et argument</h2>

<h3 style="margin-top:0; color:#854F0B;">Comparons les deux ecritures</h3>

```
def triangle(cote):       <- definition
    forward(cote)
    left(120)

triangle(100)             <- appel
```

1. Le mot `cote` apparait dans la **définition** de la fonction, entre les parentheses. On l'appelle un **paramètre**. C'est un nom que **vous choisissez** : il pourrait s'appeler `taille`, `c`, `longueur`... ce qui compte, c'est qu'il soit lisible.

2. Le nombre `100` apparait dans **l'appel** de la fonction. On l'appelle un **argument**. C'est ** la valeur** que la fonction va utiliser pour faire son travail.

3. La grande astuce du paramètre : **a chaque appel**, sa valeur peut être différente.
   - Avec `triangle(50)`, le paramètre `cote` vaut 50.
   - Avec `triangle(100)`, il vaut 100.
   - Avec `triangle(200)`, il vaut 200.

4. La définition est écrite **une seule fois**. Pourtant, elle peut produire **autant de dessins différents** qu'il y a d'appels avec des valeurs différentes. C'est exactement le contraire de ce qu'on faisait juste avant avec `triangle_petit()`, `triangle_moyen()`, `triangle_grand()`.

5. Le paramètre n'existe **que dans la fonction**. Si vous tapez `print()` apres la fonction, Python repond qu'il ne connait pas `cote`. Le paramètre est interne, comme un mot de code qu'on partage entre la définition et chaque appel.

<h3 style="color:#854F0B; margin-top:18px;">Pourquoi utiliser un paramètre ? Quatre raisons</h3>

En programmation, un paramètre n'est pas juste un confort : il repond a **quatre besoins fondamentaux**.

| # | Raison d'être | Concretement |
|---|---|---|
| 1 | **Reutilisation** | Une seule fonction peut servir dans des dizaines de contextes différents. |
| 2 | **Maintenabilite** | Si on doit changer la logique (par exemple, l'angle), on modifie en **un seul endroit** au lieu de N. Moins de risque d'oubli. |
| 3 | **Lisibilite** | `triangle(100)` se comprend immediatement. Trois lignes de `forward()` et `left()`, beaucoup moins. |
| 4 | **Composition** | Une fonction parametree peut être appelée depuis l'intérieur d'une autre fonction parametree, et ainsi de suite. C'est ce qui permet de construire des programmes complexes a partir de briques simples. |

**Vous l'avez vu en debut de seance** : `forward()`, `print()`, `range()`, `goto()`... toutes ces fonctions natives utilisent exactement le même principe. Vous savez maintenant comment elles fonctionnent, et surtout comment en creer vous-même.

## 4. Plusieurs paramètres dans une même fonction

Certaines formes ont besoin de **deux mesures** pour être completement decrites. Un **rectangle**, par exemple : il a une largeur et une hauteur.

Python permet de mettre **autant de paramètres que nécessaire** dans la définition d'une fonction. Il suffit de les separer par des **virgules**.

````python
from turtle import *

# Deux parametres : largeur et hauteur, separes par une virgule
def rectangle(largeur, hauteur):
    for _ in range(2):
        forward(largeur)
        left(90)
        forward(hauteur)
        left(90)

# Trois rectangles aux proportions differentes
# A chaque appel, l'ordre est respecte : largeur d'abord, hauteur ensuite
up()
goto(-260, -50)
down()

rectangle(100, 60)    # batiment large et bas
up()
forward(130)
down()

rectangle(60, 120)    # tour haute et etroite
up()
forward(90)
down()

rectangle(80, 80)     # batiment carre

done()
````
